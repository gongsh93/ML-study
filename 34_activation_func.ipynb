```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# hyperbolic tangent
def tanh(x):
    return list(map(lambda x : math.tanh(x), x))
```

```python
# relu
def relu(x):
    result = []
    for ele in x:
        if(ele <= 0):
            result.append(0)
        else:
            result.append(ele)
    
    return result
```

```python
x = np.linspace(-4, 4, 100)
```

```python
sig = sigmoid(x)
plt.plot(x, sig)
plt.show()
```

```python
th = tanh(x)
plt.plot(x, th)
plt.show()
```

```python
rlu = relu(x)
plt.plot(x, rlu)
plt.show()
```
